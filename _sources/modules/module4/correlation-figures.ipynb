```python
fig, ax = plt.subplots(figsize=(4,4))
ax.scatter(x,y, c='k', marker='.');
#ax.axvline(np.mean(x),c='tab:blue',linestyle='--')
#ax.text(np.mean(x)+0.05, 2, '$\overline{x}$', c='tab:blue',size=15)
ax.axhline(np.mean(y),c='tab:red',linestyle='--')
ax.text(0.4, np.mean(y)+0.15, '$\overline{y}$', c='tab:red',size=15)
for i, this_y in enumerate(y):
    if this_y < np.mean(y):
        ax.plot([x[i], x[i]], [this_y, np.mean(y)], lw=1, linestyle=':', c='tab:red')
    if this_y > np.mean(y):
        ax.plot([x[i], x[i]], [np.mean(y), this_y],  lw=1, linestyle=':', c='tab:red')
ax.set_ylabel('y')
ax.set_xlabel('x');
ax.set_xlim(0,2);
ax.set_ylim(1,7);

y_diffs = y - np.mean(y)
print(y_diffs)
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.scatter(x,y, c='k', marker='.');
ax.axvline(np.mean(x),c='tab:blue',linestyle='--')
ax.text(np.mean(x)+0.05, 2, '$\overline{x}$', c='tab:blue',size=15)
ax.axhline(np.mean(y),c='tab:red',linestyle='--')
ax.text(0.4, np.mean(y)+0.15, '$\overline{y}$', c='tab:red',size=15)
ax.set_ylabel('y')
ax.set_xlabel('x');
ax.set_xlim(0,2);
ax.set_ylim(1,7);

xy_diffs = x_diffs * y_diffs
print(xy_diffs)
sum_diffs = np.sum(xy_diffs)
print(sum_diffs)
r = sum_diffs / (np.sqrt(np.sum(x_diffs**2))*np.sqrt(np.sum(y_diffs**2)))
print(r)
```
